```python
def align_labels_with_tokens(labels, word_ids):
    new_labels = []
    current_word = None
    for word_id in word_ids:
        if word_id != current_word:
            # Start of a new word!
            current_word = word_id
            label = -100 if word_id is None else labels[word_id]
            new_labels.append(label)
        elif word_id is None:
            # Special token
            new_labels.append(-100)
        else:
            # Same word as previous token
            label = labels[word_id]
            # If the label is B-XXX we change it to I-XXX
            if label % 2 == 1:
                label += 1
            new_labels.append(label)

    return new_labels
```

```python
preds = []
golds = []
ic = 0

for i in dataset:
  ts = i['tokens']
  ls = i['ner_tags']

  if ts == []:
    continue

  inputSentence = ' '.join(ts)
  print('inputSentence', inputSentence)


  tokenized_inputs = tokenizer(inputSentence)
  word_ids = tokenized_inputs.word_ids()
  new_labels = []
  for i, wi in enumerate(word_ids):
      if i > 0 and wi != None:
        if wi < len(ls):
          new_labels.append(ls[wi])
        else:
          new_labels.append(ls[-1])

  # print(new_labels)

  output = nlp(inputSentence)

  print('output: ',output)

  # print(int(tf.argmax(outputs, axis=2)[0][0]))
  pred = [(i['index'],tags_2_i[i['entity']]) for i in output]
  predst = [0 for i in new_labels]
  for i in pred:
    if i[0] > len(predst):
      break
    predst[i[0]-1] = i[1]
  pred = predst
  # pred = [(tags_2_i[i[0]['entity']] if len(i) > 0 else 0) for i in output]
  gold = new_labels

  print('pred: ',pred)
  print('gold: ',gold)

  preds.append(pred)
  golds.append(gold)

  ic += 1
  if ic > 201:
    break
```

```python
def report(golds_bn, y_pred_bn):
  pdic = {i:0 for i in i_2_tags.keys()}
  gdict = {i:0 for i in i_2_tags.keys()}
  cpdict = {i:0 for i in i_2_tags.keys()}

  for i in golds_bn:
    for j in i:
      gdict[j] += 1

  for i in y_pred_bn:
    for j in i:
      pdic[j] += 1

  for i,v in enumerate(golds_bn):
    for j,l in enumerate(v):
      if l == y_pred_bn[i][j]:
        cpdict[y_pred_bn[i][j]] += 1

  print(cpdict)
  print(pdic)
  print(gdict)
  res = 0
  for i in i_2_tags.keys():
    if gdict[i] != 0:  # i != 0 and
      prec = 0 if pdic[i] == 0 else (cpdict[i]/pdic[i])
      rec = 0 if gdict[i] == 0 else (cpdict[i]/gdict[i])
      res += 0 if rec+prec == 0 else (2*prec*rec)/(rec+prec)

  return res/len(i_2_tags.keys())
```
